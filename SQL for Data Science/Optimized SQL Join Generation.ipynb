### Activity 2:  Basic Join Query 

Now that we understand our data, let's write a basic join query to combine customer and transaction information.

<b>Step 1:</b> Write a query to join the customers and transactions tables:

```python
query = """
SELECT c.customer_id, c.join_date, t.transaction_id, t.sale_price
FROM customers c
JOIN transactions t ON c.customer_id = t.customer_id
LIMIT 10;
"""
df = pd.read_sql_query(query, conn)
print("Joined Customer and Transaction Data:")
display(df)
```

<b>Step 2:</b> Modify the query to include more relevant columns and add a WHERE clause:

```python
query = """
SELECT c.customer_id, c.join_date, c.preferred_store, t.transaction_id, t.sale_price, t.store_location
FROM customers c
JOIN transactions t ON c.customer_id = t.customer_id
WHERE c.is_member = 1
LIMIT 10;
"""
df = pd.read_sql_query(query, conn)
print("Joined Data for Members:")
display(df)
```

### Activity 3: AI-Assisted Query Optimization 

Let's use AI to help us generate an optimized SQL join query for a more complex analysis.

<b>Step 1:</b> Prepare the prompt for the AI. We‚Äôve provided a sample prompt you can use below. The prompt contains the column and data type information for two tables called ‚Äúcustomers‚Äù and ‚Äútransactions‚Äù in a SQLite database.


Given two tables in a SQLite database:

1. customers:
   - customer_id (string)
   - join_date (date)
   - is_member (integer, 0 or 1)
   - zip_code (integer)
   - birth_year (integer)
   - preferred_store (string)

2. transactions:
   - transaction_id (string)
   - date_time (datetime)
   - store_location (string)
   - customer_id (string)
   - book_id (string)
   - sale_price (float)
   - payment_method (string)
   - is_online (integer, 0 or 1)

Generate an optimized SQL join query to list customer IDs along with their total order amounts, 
focusing on in-store purchases by members who joined in 2022. Include their preferred store and 
actual purchase location. Order the results by total order amount descending.


<b>Step 2:</b> Use an AI assistant (like ChatGPT) to generate the query. For example, here is an AI response : 

```python
ai_response = """
Here's an optimized SQL join query based on your requirements:

SELECT 
    c.customer_id,
    c.preferred_store,
    SUM(t.sale_price) AS total_order_amount,
    COUNT(DISTINCT t.transaction_id) AS transaction_count,
    GROUP_CONCAT(DISTINCT t.store_location) AS purchase_locations
FROM 
    customers c
JOIN 
    transactions t ON c.customer_id = t.customer_id
WHERE 
    c.is_member = 1
    AND strftime('%Y', c.join_date) = '2022'
    AND t.is_online = 0
GROUP BY 
    c.customer_id, c.preferred_store
ORDER BY 
    total_order_amount DESC
LIMIT 20;

This query:
1. Joins the customers and transactions tables.
2. Filters for members who joined in 2022 and made in-store purchases.
3. Calculates the total order amount and transaction count for each customer.
4. Includes the preferred store and a list of actual purchase locations.
5. Orders the results by total order amount descending.
6. Limits the output to 20 rows for performance.

The use of GROUP_CONCAT for purchase locations provides a compact way to see all locations where a customer made purchases.
"""

print("AI-Generated Query:")
print(ai_response)
```
